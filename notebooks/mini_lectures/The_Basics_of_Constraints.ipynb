```python
#@title Imports
import jax 
import jax.numpy as jnp
from functools import partial 
import matplotlib as mpl
from matplotlib import pyplot as plt
from matplotlib import rcParams
rcParams['image.interpolation'] = 'None'
rcParams['image.cmap'] = 'viridis'
rcParams['axes.grid'] = False
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
#@title **Choice Set: $\mathcal{X}$**
xp=jnp.linspace(0.,6.,200)  # N
yp=jnp.linspace(0.,12.,200)  # M

X, Y = jnp.meshgrid(xp, yp)

def func(x,y):
  return 0.*x*y

Z = func(X,Y)
plt.contourf(xp,yp,Z) 
plt.title('Apples', loc='left', size=14)
plt.xlabel('Oranges', size=14)
plt.show()
```

```python
#@title **Constraint Function: $F_p: \mathcal{X} \to \{0,1\} $**
def F(income, po, pa, xo, xa):
  return jnp.where(po*xo + pa*xa > income, 1., 0.)

income, po, pa = 20., 5., 2. 

print(f"The bundle (2,1) is {'feasible' if F(income, po, pa, 2, 1) ==0 else 'not feasible'}")
print(f"The bundle (10,10) is {'feasible' if F(income, po, pa, 10, 10) ==0 else 'not feasible'}")
```

```python
#@title **Feasible Set: $F_p^{-1}(0) \subset \mathcal{X}$**
xp=jnp.linspace(0.,6.,200)  # N
yp=jnp.linspace(0.,12.,200)  # M

X, Y = jnp.meshgrid(xp, yp)

Z = partial(F, income, po, pa)(X,Y)
plt.contourf(xp,yp,Z) 
plt.title('Apples', loc='left', size=14)
plt.xlabel('Oranges', size=14)
plt.show()
```
